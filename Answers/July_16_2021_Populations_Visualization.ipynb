```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
from scipy.stats import norm
```

```python
from IPython.display import Image
pwd = os.getcwd()
Image(pwd + '\\July_16_2021_Questions.png')
```

```python
s = np.random.normal(loc=50, scale=12.5, size=2500)
s = sorted(s)
```

```python
pdf = norm.pdf(s, np.mean(s), np.std(s))
pdf
```

```python
plt.figure(figsize=(15,5))
plt.xlabel('Income')
plt.ylabel('Normal Distribution')
plt.title('Population Wealth',fontdict={
    'family': 'monospace',
    'color':  '#000066',
    'weight': 'bold',
    'size': 16})
plt.plot(s,pdf)
```
