```python
inputA = input('Type number seperated by space:')
inputA = inputA.split(' ')
for i in range(0,len(inputA)):
    inputA[i] = int(inputA[i])
inputint = int(input('size? '))

lst1 = inputA[0:inputint]
output = []
for i in range(len(inputA) + 1 - inputint):
    output.append(max(lst1))
    del lst1[0]
    if i < len(inputA) - inputint:
        lst1.append(inputA[i+inputint])
print(output)
```
