```python
births_deaths = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_births.csv')
pop = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_populations.csv')
```

```python
output = pd.DataFrame(
    {'State' : pop['State'],
    'Population' : pop['Population'],
    'Births' : births_deaths['Births'],
    'Birth Rate' : births_deaths['Births'] / pop['Population'],
    'Deaths' : births_deaths['Deaths'],
    'Death Rate' : births_deaths['Deaths'] / pop['Population']}
)

output['Births - Deaths'] = output['Births'] - output['Deaths']
output['Net Population Change'] = output['Population'] + output['Births'] - output['Deaths']
output['Projected Pop in 5 years'] = output['Population'] + (output['Births'] - output['Deaths']) * 5

output.style.format(
    {'Population': '{:,}',
    'Births': '{:,}',
    'Deaths': '{:,}', 
    'Net Population Change': '{:,}',
    'Births - Deaths' : '{:,}',
    'Projected Pop in 5 years' : '{:,}'}
)
```

```python
def color_negative_red(value):
  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'green'
  else:
    color = 'black'
  return 'color: %s' % color
```

```python
output.style.applymap(color_negative_red, subset='Births - Deaths')
```
