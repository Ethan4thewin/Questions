```python
s = np.random.normal(loc=50, scale=12.5, size=2500).round(2)
s = pd.Series(s)
```

```python
for i in range(500):
    a = []
    for y in range(2):
        a.append(s[random.randint(0,len(s)-1)])
    if a[1] == a[0]:
        a[1] = s[random.randint(0,len(s)-1)]
    else: pass
    pot = sum(a)

    b = []
    b.append(random.uniform(0,pot))
    b.append(pot - b[0])
    s.replace({a[0]:b[0], a[1]:b[1]},inplace=True)
```
