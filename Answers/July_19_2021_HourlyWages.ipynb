```python
import pandas as pd
import numpy as np
```

```python
hours = np.random.uniform(38,43,size=50)
wage = np.random.uniform(21,60,size=50)
```

```python
dataset = pd.DataFrame(data={'Weekly Hours':hours, 'Hourly Wages' :wage})
dataset['Pre-tax Earnings'] = dataset['Weekly Hours'] * dataset['Hourly Wages']
dataset['Tax Rates'] = np.take([.15,.30], np.digitize(dataset['Pre-tax Earnings'], bins=[2_000]))
dataset['Post-tax Earnings'] = dataset['Pre-tax Earnings'] * (1-dataset['Tax Rates'])
dataset.head().style.format('{:.2f}')
```
