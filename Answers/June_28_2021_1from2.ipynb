```python
input1 = input('Type number seperated by space:')
input1 = input1.split(' ')
for i in range(0,len(input1)):
    input1[i] = int(input1[i])

all_multiplied = list()

for i in range(0,len(input1)):
    for z in range(i,len(input1)):
        if i == z : pass
        else:
            all_multiplied.append(input1[i] * input1[z])

result = [i for i in all_multiplied if i in input1]

if result != 0:
    print(max(result))
else: print('N/A')
```
