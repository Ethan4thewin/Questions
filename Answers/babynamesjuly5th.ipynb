```python
import os
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/ddi_baby_names.csv')
dataset
```

```python
dataset['name'].nunique()
```

```python
dataset_boy = dataset[dataset['gender'] == 'M'].reset_index()
dataset_girl = dataset[dataset['gender'] == 'F'].reset_index()
print('boy ',dataset_boy['name'].nunique(),'\ngirl ',dataset_girl['name'].nunique())
```

```python
dataset_boy.head(10)
```

```python
namecount_boy = dataset_boy.groupby('name')['count'].sum().reset_index()
namecount_boy.nlargest(10,'count')
```

```python
namecount_girl = dataset_girl.groupby('name')['count'].sum().reset_index()
namecount_girl.nlargest(10,'count')
```

```python
dataset_2010 = dataset[dataset['year'] >= 2010]
namecount_2010 = dataset_2010.groupby('name')['count'].sum().reset_index()
namecount_2010.nlargest(10,'count')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def growth(lst1):
    for n in lst1:
        namecount = dataset_2010[dataset_2010['name'] == n].groupby('year')['count'].sum().reset_index()
        plt.plot(namecount['year'],namecount['count'],marker='o',ms=2.5)
```

```python
growth(['Emma',
 'Sophia',
 'Olivia',
 'Noah',
 'Isabella',
 'Liam',
 'Jacob',
 'Mason',
 'William',
 'Ava'])
plt.grid()
plt.plot()
```
